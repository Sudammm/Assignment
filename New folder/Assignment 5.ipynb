```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
df['target']=iris.target
```

```python
df
```

```python
#cross_val_score
help(LogisticRegression())
```

```python
#Logistics Regression
score_log=np.mean(cross_val_score(LogisticRegression(solver="liblinear", multi_class="ovr"),iris.data,iris.target,cv=10))
score_log
```

```python
#SVM
score_svm=np.mean(cross_val_score(SVC(gamma='auto'),iris.data,iris.target,cv=10))
score_svm
```

```python
#DecisionTree
from sklearn.tree import DecisionTreeClassifier
score_tree=np.mean(cross_val_score(DecisionTreeClassifier(),iris.data,iris.target))
score_tree
```

```python
# Random forest
score_forest=np.mean(cross_val_score(RandomForestClassifier(n_estimators=40),iris.data,iris.target,cv=10))
score_forest
```

```python
print(f'score_log:{score_log},\nscore_SVM:{score_svm},\nscore_tree:{score_tree},\nscore_forest:{score_forest}')
```
